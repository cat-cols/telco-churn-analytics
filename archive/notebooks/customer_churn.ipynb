```python
# Drop Null amounts
data = data.dropna()
```

```python
features = data.iloc[:,0:-1]
```

```python
labels = data.iloc[:,-1]
```

```python
features.head(5)
```

```python
labels.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features,
                 labels,
                 test_size=0.2,
                 random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numerical_columns = ['Age','Tenure','Usage Frequency','Support Calls',
                     'Payment Delay','Total Spend',
                     'Last Interaction']
```

```python
X_train_numerical = X_train[numerical_columns]
```

```python
#scaler.fit() computes mean & StdDev of each column
#scaler.transform() uses mean & StdDev to scale values of each column
scaler = StandardScaler()
```

```python
scaler.fit(X_train_numerical)
```

```python
X_train_scaled = scaler.transform(X_train_numerical)
```

```python
X_train_scaled_df = pd.DataFrame(data=X_train_scaled,
                                 index=X_train.index,
                                 columns=numerical_columns)
```

```python
X_train_scaled_df.head(5)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
categorical_columns = ['Subscription Type', 'Contract Length']
```

```python
X_train_categorical = X_train[categorical_columns]
```

```python
encoder = OneHotEncoder()
```

```python
# encoder.fit():
# 1. checks unique values within each category
# 2. prepares labels of output column
encoder.fit(X_train_categorical)
```

```python
X_train_encoded = encoder.transform(X_train_categorical)
```

```python
print(type(X_train_encoded))
```

```python
print(X_train_encoded)
```

```python
X_train_encoded.todense()
```

```python
encoder.get_feature_names_out()
```

```python
X_train_encoded_df = pd.DataFrame(X_train_encoded.todense(),
                                  index=X_train.index,
                                  columns= encoder.get_feature_names_out())
```

```python
X_train_encoded_df.head(10)
```

```python
X_train_transf = pd.concat([X_train['CustomerID'],
                            X_train_scaled_df,
                            X_train_encoded_df], axis=1)
```

```python
X_train_transf.head(5)
```

```python
#scaling the numerical columns in the testing set
X_test_scaled = scaler.transform(X_test[numerical_columns])
```

```python
# scaling the numerical columns in the testing set
X_test_scaled = scaler.transform(X_test[numerical_columns])
X_test_scaled_df = pd.DataFrame(data = X_test_scaled,
                                index=X_test.index,
                                columns=numerical_columns)
```

```python
# encoding the categorical columns in testing set
X_test_encoded = encoder.transform(X_test[categorical_columns]) 
X_test_encoded_df = pd.DataFrame(data=X_test_encoded.todense(),
                                 index=X_test.index,
                                 columns= encoder.get_feature_names_out())
```

```python
# concat the two DataFrames
X_test_trans = pd.concat([X_test['CustomerID'],
                          X_test_scaled_df,
                          X_test_encoded_df], axis=1)
```

```python
X_test_trans.head(5)
```

```python
X_train_transf.to_parquet("train.parquet")
X_test_trans.to_parquet("test.parquet")
```
